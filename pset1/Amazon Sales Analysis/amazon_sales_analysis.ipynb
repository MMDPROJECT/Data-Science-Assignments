### Initial Data Preprocessing


Now we need to perform an Initial Data Preprocessing in order to be able to work with the data at our initial stages of work. We would complete this process, later in **Data Preprocessing** Section.

Based on what we saw earlier in our early stages of **Exploratory Data Analysis**, we decide to apply basic transformation to change the data type of the columns **discounted_price**, **actual_price**, **discount_percentage**, **rating** and **rating_count** from *object* to *numeric*. We would also drop the rows which have missing values without any further analysis, simply because there are only 2 rows of such. Also, we drop the other columns that we don't want to use in our entire analysis.

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'].apply(lambda x:x[1:].replace(',', '')), errors= 'coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'].apply(lambda x:x[1:].replace(',', '')), errors= 'coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].apply(lambda x:x[:len(x) - 1]), errors= 'coerce')
df['rating'] = pd.to_numeric(df['rating'], errors= 'coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].apply(lambda x:x.replace(',', '') if isinstance(x, str) else x), errors= 'coerce')
df.dropna(inplace= True)
df = df[['product_id', 'category', 'discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']]
df.info()
```

# Hypothesis Testing:

1. Does the discounted price significantly impact product ratings?

- Hypothesis Test & Why: Based on what we saw earlier, in Exploratory Data Analysis(EDA) section, We choose ***Spearman Correlation Test*** for testing this hypothesis, since the two variables were both skewed to one of their ends and were not normally distributed, Also there has been many outliers in the scatter plot of these two variables. And also as we know ***Pearson Correlation Test*** has all of these conditions as its assumptions, thus we can't use it.

- Result: From hypothesis testing we get Spearman Correlation: 0.080, P-value: 0.002.

- Interpretation & Discussion: The p-value is in contradictory to what we would expect it to be, because as we saw earlier from the scatter plot of these two variables, they don't seem to be correlated. Thus we consider the p-value as invalid and conclude from what we have analysed from earlier sections that these two variables are not significantly correlated.

```python
spearman_corr, p_value = stats.spearmanr(df[['discounted_price', 'rating']])
print(f"Spearman Correlation: {spearman_corr:.3f}")
print(f"P-value: {p_value:.3f}")
```

2. Are product categories and high/low ratings independent?

- Hypothesis Test: We use ***Chi-Squared Independence Test***. 

- Considerations: We have dropped the rows which are in a **category** which has occured less than or equal to 5 times in the entire dataseet. We utilize this police, simply because many of the **categories** have just occured once in our dataset and analysis of **rating** withing their group is some what meaningless. One last note is that we took **ratings** above or equal to 4 as high and the others as low.

- Result: performing the hypothesis test, we get Chi-Squared Statistic: 213.369, P-value: 0.000.

- Interpretation & Discussion: We have obtained an extremely low p-value which totally matches with our prior analysis in Exploratory Data Analysis(EDA) section. Thus we can say that there is a significant difference between different **categories** in terms of being in different **rating** levels.

```python
# Get category counts
category_counts = df['category'].value_counts()

# Get categories that appear more than 5 times
categories_to_keep = category_counts[category_counts > 5].index
# print(category_counts[category_counts > 5].index)

# Filter the DataFrame to keep only these categories
df_filtered = df[df['category'].isin(categories_to_keep)]

contingency_table = pd.crosstab(df_filtered['category'], df_filtered['rating'].apply(lambda x: 'high' if x >= 4 else 'low'))

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Squared Statistic: {chi2:.3f}")
print(f"P-value: {p:.3f}")
print(f"Degrees of Freedom: {dof}")
```

3. Is there a significant difference in ratings between high-discount and low-discount products?

- Hypothesis Test & Why: We use ***student's two sample t-test***. Based on our earlier analysis on the distributions of these two variables, we saw that they are both normally distributed and have the same variances, thus this choice seemed very reasonable. 

- Considerations: We have taken **discount_percentages** greater or equal to 50% as high and the others and low.

- Result: performing the hypothesis test, we get T-statistic: -4.261, P-value: 0.000.

- Interpretation & Discussion: The obtained p-value is very small, which actually contradicts with what we have got to in our analysis. By having a closer look at the kde-plot of these two groups, which is pictured below, we can see that our analysis was wrong and mean of the **rating** in low-discount sales is actually a little bit greater than the high-discount group. Thus we reject the null hypothesis and conclude that there is a significant difference between these two groups. 

```python
temp_df = df.copy()
temp_df['discount_percentage'] = temp_df['discount_percentage'].apply(lambda x: 'high-discount' if x >= 50 else 'low-discount')
high_discount_group = temp_df[temp_df['discount_percentage'] == 'high-discount']['rating']
low_discount_group = temp_df[temp_df['discount_percentage'] == 'low-discount']['rating']
sns.kdeplot(temp_df, x='rating', hue='discount_percentage')
plt.tight_layout()
output_path = 'distro_of_rating_for_different_discount_percentages.png'
plt.savefig(output_path)
plt.show()
t_stat, p_value = stats.ttest_ind(high_discount_group, low_discount_group)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")
```

4. Do more expensive products receive higher ratings on average?

- Hypothesis Test & Why: We use ***ANOVA one-way Test***. Based on the our earlier analysis in Exploratory Data Analysis(EDA) section, we saw that the distribution of all of the three different groups of **actual_price**, namely, low, medium and high are normally distributed, thus we can say the assumptions of the test hold and we can go for it.   

- Considerations: We have splitted the **actual_price** into three different levels based on 33rd and 66th percentiles.

- Result: performing the hypothesis test, we get F-statistic: 0.907, P-value: 0.404.

- Interpretation & Discussion: p-value is very high and this actually matches to what we have concluded from our earlier analysis. Thus we conclude that there is no significant difference between different levels of **actual_price** in terms of **rating**.

```python
# Calculate percentiles
low_threshold = df['actual_price'].quantile(0.33)   # 33rd percentile
high_threshold = df['actual_price'].quantile(0.66)  # 66th percentile

# Assign categories based on percentiles
temp_df = df.copy()
temp_df['price_category'] = df['actual_price'].apply(lambda x: 'low' if x <= low_threshold else 'medium' if x <= high_threshold else 'high')
low_price_group = temp_df[temp_df['price_category'] == 'low']['rating']
medium_price_group = temp_df[temp_df['price_category'] == 'medium']['rating']
high_price_group = temp_df[temp_df['price_category'] == 'high']['rating']

# Perform ANOVA
f_stat, p_value = stats.f_oneway(low_price_group, medium_price_group, high_price_group)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.3f}")
```

5. Does the distribution of rating counts follow a normal distribution?

- Hypothesis Test: We use ***Shapiro-Wilk Test***. 

- Result: performing the hypothesis test, we get Shapiro-Wilk Test Statistic: 0.407, P-value: 0.000.

- Interpretation & Discussion: P-value is exteremly low, which matches with our prior analysis as well. Thus we conclude that the **rating_count** is not normally distributed.

```python
temp_df = df.copy()
max_rating_count_per_product_series = temp_df.groupby('product_id')['rating_count'].max()
max_rating_count_per_product_series.name = 'max_rating_count'
max_rating_count_per_product_series.shape
# Perform the Shapiro-Wilk test
statistic, p_value = stats.shapiro(max_rating_count_per_product_series)

# Print the result
print(f"Shapiro-Wilk Test Statistic: {statistic:.3f}")
print(f"P-value: {p_value:.3f}")
```
